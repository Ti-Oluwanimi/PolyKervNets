```python
#datasizecombined
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
dt = pd.read_csv("/home/toluwani/Documents/PolyKervNets/Saved/Data - Sheet1.csv")
dt.head(30)
```

```python
sns.set(rc={'figure.figsize':(12,10)})
with sns.axes_style("whitegrid"):
    sns.scatterplot(data=dt, x="Multiplicative Depth", y="Top 1 Accuracy% ", hue="Variant", style="Model", markers=["s","D","^","d","X","P"], palette='deep', s=250, legend=True)
    sns.lineplot(data=dt, x="Multiplicative Depth", y="Top 1 Accuracy% ", hue="Variant", linewidth = 2.0, legend=False)
    plt.legend(loc='lower right', fontsize=16, markerscale=2.0)
```
